```python
from nptdms import TdmsFile
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
from mpl_toolkits.axes_grid1 import make_axes_locatable

from HyperspectralAnalysis import HyperspectralAnalysis 
hyps = HyperspectralAnalysis()

plt.rcParams['figure.facecolor'] = 'none'
```

```python
def process_data(path_int, pathdc, pathbc, path_positions):
    pixel_params, wave_raw, inten_raw = hyps.read_tdms(tdms_file=TdmsFile.read(path_int), datalen=670)
    _, w1, dc_raw =     hyps.read_tdms(tdms_file=TdmsFile.read(pathdc), datalen=1340)
    _, w2, bc_raw =    hyps.read_tdms(tdms_file=TdmsFile.read(pathbc),  datalen=1340)
    wc_minus_dc = hyps.process_wcdc(dc_raw=dc_raw, 
                         bc_raw=bc_raw,
                         pixel_params=pixel_params)
    positions = np.loadtxt(path_positions)
    numPart = len(positions)
    return numPart, positions, wave_raw, inten_raw, wc_minus_dc

def findNPs(numPart, positions, inten_raw,):
    %matplotlib notebook
    for i in range(numPart):
        plt.annotate(str(i), fontsize=14, xy=[int(positions[i][0]+5), int(positions[i][1]+5)],color='red')
    plt.imshow(np.sum(inten_raw,2).T, cmap='turbo')
```

```python
# For Mg NPs, Jan. 28
wavei = 50 # index of starting wavelength 
wavef = 670 # index of end wavelength
pathbc = '../Mg Analyses/250128/mg_recWC3.tdms'
pathdc = '../Mg Analyses/250128/mg_recDC6.tdms'


path_int1 = '../Mg Analyses/250128/mg_roi3_4.tdms'
path_positions1 = '../Mg Analyses/250128/mg_roi3_3/positions.txt'


# path_int2 = '../Mg Analyses/250202_first/roi3_g1_2.tdms'
# path_positions2 = '../Mg Analyses/250202_first/positions_roi3_2.txt'
# foc2 = 19
# path_int3 = '../Mg Analyses/250202_first/roi3_g1_3.tdms'
# path_positions3 = '../Mg Analyses/250202_first/positions_roi3_3.txt'
# foc3 = 12

```

```python
numPart1, positions1, wave_raw1, inten_raw1, wc_minus_dc1 = process_data(path_int1,
                                                                         pathdc, 
                                                                         pathbc, 
                                                                         path_positions1)

numPart2, positions2, wave_raw2, inten_raw2, wc_minus_dc2 = process_data(path_int2,
                                                                         pathdc, 
                                                                         pathbc, 
                                                                         path_positions2)

numPart3, positions3, wave_raw3, inten_raw3, wc_minus_dc3 = process_data(path_int3,
                                                                         pathdc, 
                                                                         pathbc, 
                                                                         path_positions3)
```

```python
def calc_all_DFS(wave_raw, inten_raw, numPart, positions, wc_minus_dc):
    wave = wave_raw[wavei:wavef]
    gam_tot = np.zeros(numPart)
    eres_tot = np.zeros(numPart)  
    fit_tot = np.zeros((wavef-wavei, numPart))

    dfs_tot = hyps.calc_DFS_localgrid(specfin=inten_raw,
                        numPart=numPart,
                        positions=positions,
                        wavei=wavei,
                        wavef=wavef,
                        wc_minus_dc=wc_minus_dc,
                        offset=4)

    for npi in range(0, numPart):
        DFSi = dfs_tot[:, npi]
        fit_tot[:,npi], popt = hyps.fit_spectrum(wave, DFSi)
        gam_tot[npi] = popt[1]; eres_tot[npi] = popt[2]
        
    return wave, dfs_tot, fit_tot, gam_tot, eres_tot
```

```python
wave1, dfs_tot1, fit_tot1, gam_tot1, eres_tot1 = calc_all_DFS(wave_raw=wave_raw1, 
                                             inten_raw=inten_raw1, 
                                             numPart=numPart1, 
                                             positions=positions1,
                                             wc_minus_dc=wc_minus_dc1)

wave2, dfs_tot2, fit_tot2, gam_tot2, eres_tot2 = calc_all_DFS(wave_raw=wave_raw2, 
                                             inten_raw=inten_raw2, 
                                             numPart=numPart2, 
                                             positions=positions2,
                                             wc_minus_dc=wc_minus_dc2)

wave3, dfs_tot3, fit_tot3, gam_tot3, eres_tot3 = calc_all_DFS(wave_raw=wave_raw3, 
                                             inten_raw=inten_raw3, 
                                             numPart=numPart3, 
                                             positions=positions3,
                                             wc_minus_dc=wc_minus_dc3)
```

```python
numcol = numPart1
colors=['tab:blue', 'tab:orange', 'tab:green']

fig, ax = plt.subplots(3, numcol, figsize=[numcol*1.7, 3*2.4])

for npi in range(0,numPart1):
    ax[0,npi].plot(wave1, dfs_tot1[:,npi], color=colors[0])
    ax[0,npi].plot(wave1, fit_tot1[:,npi], color='k', linestyle='--')
    titlename1 = ('NP: ' + str(npi) + ' \n' + 
                 str('$E_{res}=$ ') + str(int(np.round(eres_tot1[npi]))) + ' nm \n' + 
                 str('$\Gamma=$ ') + str(int(np.round(gam_tot1[npi])))+str(' nm'))
    ax[0, npi].set_title(titlename1, y=.8, fontsize=10)
    
    ax[1,npi].plot(wave2, dfs_tot2[:,npi], color=colors[1])
    ax[1,npi].plot(wave2, fit_tot2[:,npi], color='k', linestyle='--')
    titlename2 = (str('$E_{res}=$ ') + str(int(np.round(eres_tot2[npi]))) + ' nm \n' + 
                  str('$\Gamma=$ ') + str(int(np.round(gam_tot2[npi])))+str(' nm'))
    ax[1, npi].set_title(titlename2, y=.8, fontsize=10)

    ax[2,npi].plot(wave3, dfs_tot3[:,npi], color=colors[2])
    ax[2,npi].plot(wave3, fit_tot3[:,npi], color='k', linestyle='--')
    titlename3 = (str('$E_{res}=$ ') + str(int(np.round(eres_tot3[npi]))) + ' nm \n' + 
                  str('$\Gamma=$ ') + str(int(np.round(gam_tot3[npi])))+str(' nm'))
    ax[2, npi].set_title(titlename3, y=.8, fontsize=10)

    
    ax[0,npi].set_xlim([500,900]);
    ax[0,npi].set_ylim([0, 1.3]);
    ax[0,npi].set_yticks([0,1.]);

    ax[2,npi].set_xticks([500, 750, ])
    ax[1,npi].set_xlim([500,900]);

    ax[1,npi].set_ylim([0,1.3]);
    ax[1,npi].set_yticks([0,1]);

    ax[2,npi].set_ylim([0, 1.3]);
    ax[2,npi].set_yticks([0,1]);
    ax[2,npi].set_xlim([500,900]);

    if npi > 0:
        ax[0,npi].set_xticklabels([])
        ax[0,npi].set_yticklabels([])
        ax[1,npi].set_xticklabels([])
        ax[1,npi].set_yticklabels([])
        ax[2,npi].set_xticklabels([])
        ax[2,npi].set_yticklabels([])
        

ax[1,0].set_xlabel('Wavelength [nm]')
ax[1,0].set_ylabel('DFS [unitless]')
fig.subplots_adjust(wspace=0, hspace=0)
# # fig.savefig('auNRs_subset_allspectra.png', dpi=600)
```
