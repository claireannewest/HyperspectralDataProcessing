```python
from nptdms import TdmsFile
import matplotlib.pyplot as plt
import scipy.io
import numpy as np
```

```python
from HyperspectralAnalysis import HyperspectralAnalysis 
hyps = HyperspectralAnalysis()
```

### Read in data files

```python
path_int = '../Mg Analyses/241221/nps_in_glycerol/mg110_glyc_33.tdms'
pathbc = '../Mg Analyses/241221/nps_in_glycerol/mg110_glyc_WC34.tdms'
pathdc = '../Mg Analyses/241221/nps_in_glycerol/mg110_glyc_DC35.tdms'

pixel_params, wave_raw, inten_raw = hyps.read_tdms(tdms_file=TdmsFile.read(path_int), datalen=670)
_, w1, dc_raw =     hyps.read_tdms(tdms_file=TdmsFile.read(pathdc), datalen=1340)
_, w2, bc_raw =    hyps.read_tdms(tdms_file=TdmsFile.read(pathbc),  datalen=1340)
```

### Process the data


```python
wc_minus_dc = hyps.process_wcdc(dc_raw=dc_raw, 
                         bc_raw=bc_raw,
                         pixel_params=pixel_params)

bk_global, indices = hyps.back_global(I_raw=inten_raw,
                                 percent=0.1)

DFS_backglob = hyps.calc_DFS(inten_raw=inten_raw, 
                             wc_minus_dc=wc_minus_dc, 
                             back=bk_global,
                             )
```

### Check new python code against matlab for a single NP


```python
npi = 4  # NP index

# Finish calculating DFS for npi in python implementation 
positions = np.loadtxt('../Mg Analyses/241221/nps_in_glycerol/mg110_glyc_33/positions.txt')
xi = int(positions[npi,0])-1
yi = int(positions[npi,1])-1
sum_DFS = hyps.avg3pixels(M=DFS_backglob, xi=xi, yi=yi)
sum_DFS_short = sum_DFS[99:]
wave_short = wave_raw[99:]

# Determine DFS for npi from matlab
mat = scipy.io.loadmat('../Mg Analyses/241221/nps_in_glycerol/mg110_glyc_33/all_spectra_mg110_glyc_33.mat')
allSpec = np.round(mat['allSpec'],5)
waves_mat = mat['rawwvlths'].reshape(571,1)

# Plot up image and spectrum
fig, ax = plt.subplots(1,2, figsize=[6,3.5])
ax[0].imshow(np.sum(DFS_backglob,2).T, cmap='turbo')
ax[0].add_patch(plt.Circle([xi, yi], radius=7, edgecolor='r',lw=1, fill=False))        
ax[1].plot(wave_short, sum_DFS_short, label='reprod. in python')
ax[1].plot(waves_mat, allSpec[:,npi], label='matlab', alpha=.5)

ax[1].set_xlim([500,900])
ax[1].legend(frameon=False);
```

```python
print('Maximum error: ', np.round(np.max(abs(sum_DFS_short-allSpec[:,npi])), 5))
```
