```python
from nptdms import TdmsFile
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from matplotlib.pyplot import cm
from mpl_toolkits.axes_grid1 import make_axes_locatable
import scipy.io
import numpy as np
```

```python
from HyperspectralAnalysis import HyperspectralAnalysis 
hyps = HyperspectralAnalysis()
```

### Read in data files

```python
path_int = '../data_from_2024-12-21/mg110_glyc_33.tdms'
pathbc = '../data_from_2024-12-21/mg110_glyc_WC34.tdms'
pathdc = '../data_from_2024-12-21/mg110_glyc_DC35.tdms'

bot_pix, top_pix, wave_raw, inten_raw = hyps.read_tdms(tdms_file=TdmsFile.read(path_int), datalen=670)
_, _, w1, dc_raw =     hyps.read_tdms(tdms_file=TdmsFile.read(pathdc), datalen=1340)
_, _, w2, bc_raw =    hyps.read_tdms(tdms_file=TdmsFile.read(pathbc),  datalen=1340)
```

### Process the data


```python
stan = hyps.process_wcdc(dc_raw=dc_raw, 
                         bc_raw=bc_raw)
bkavg = hyps.calc_background_avg(I_raw=inten_raw,
                                 percent=0.1)
specfin = hyps.calc_DFS_onepixel(inten_raw=inten_raw, 
                                 stan=stan, 
                                 bkavg=bkavg,
                                 top_pix=top_pix,
                                 bot_pix=bot_pix)
```

### Check new python code against matlab for a single NP


```python
npi = 10  # NP index

# Finish calculating DFS for npi in python implementation 
positions = np.loadtxt('../data_from_2024-12-21/mg110_glyc_33/positions.txt')
xi = int(positions[npi,0])-1
yi = int(positions[npi,1])-1
sum_DFS = hyps.calc_DFS_avg3pixels(specfin=specfin, xi=xi, yi=yi)
sum_DFS_short = sum_DFS[99:]
wave_short = wave_raw[99:]

# Determine DFS for npi from matlab
mat = scipy.io.loadmat('../data_from_2024-12-21/mg110_glyc_33/all_spectra_mg110_glyc_33.mat')
allSpec = np.round(mat['allSpec'],5)
waves_mat = mat['rawwvlths'].reshape(571,1)

# Plot up image and spectrum
fig, ax = plt.subplots(1,2, figsize=[6,3.5])
ax[0].imshow(np.sum(specfin,2).T, cmap='turbo')
ax[0].add_patch(plt.Circle([xi, yi], radius=7, edgecolor='r',lw=1, fill=False))        
ax[1].plot(wave_short, sum_DFS_short, label='reprod. in python')
ax[1].plot(waves_mat, allSpec[:,npi], label='matlab', alpha=.5)

ax[1].set_xlim([500,900])
ax[1].legend(frameon=False);
```

```python
print('Maximum error: ', np.round(np.max(abs(sum_DFS_short-allSpec[:,npi])), 5))
```
